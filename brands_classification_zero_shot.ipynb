## Downloading and Setting up Model Pipeline

```python
from transformers import pipeline
classifier = pipeline("zero-shot-classification",
                      model="facebook/bart-large-mnli", device=0)
```

## Example_1

```python
sequence_to_classify = "one day I will see the world"
candidate_labels = ['travel', 'cooking', 'dancing']
classifier(sequence_to_classify, candidate_labels)
#{'labels': ['travel', 'dancing', 'cooking'],
# 'scores': [0.9938651323318481, 0.0032737774308770895, 0.002861034357920289],
# 'sequence': 'one day I will see the world'}
```

## Example_2

```python
candidate_labels = ['travel', 'cooking', 'dancing', 'exploration']
classifier(sequence_to_classify, candidate_labels, multi_label=True)
#{'labels': ['travel', 'exploration', 'dancing', 'cooking'],
# 'scores': [0.9945111274719238,
#  0.9383890628814697,
#  0.0057061901316046715,
#  0.0018193122232332826],
# 'sequence': 'one day I will see the world'}
```

### All Cases Output

```python
# df.sample(1)['desc_translate'].values[0]
df_clean.iloc[0]['desc_translate']
```

```python
df_ss = df_clean.sample(frac=1)
print(df_ss.shape)
```

```python
df_ss.shape[0]*10
```

```python
a=[]
for i in range (0, df_ss.shape[0]):
  text = df_ss.iloc[i]['desc_translate']
  # print(text)

  candidate_labels = ['fashion accessories', 'clothing', 'home decor', 'cosmetics', 'wellness', 'food', 'flowers', 'gifting']
  sequence_to_classify=[text]

  result_list = classifier(sequence_to_classify, candidate_labels, multi_label=True)
  result = pd.DataFrame({k: v for element in result_list for k, v in element.items() if isinstance(element, dict)})
  a.append(result)

df_results = pd.concat(a)
df_results.shape
```

```python
df_results
```

```python
pd.pivot_table(df_results, index='sequence',columns='labels',values='scores',aggfunc=np.sum).to_csv("check_results.csv")
```
