### Function for word_token counts

```python
import pandas as pd


def tokenize_and_count_words(df, column_name):
  """Tokenizes each word available in a column of a Pandas DataFrame and counts the frequency of each word token.

  Args:
    df: A Pandas DataFrame.
    column_name: The name of the column containing the text values.

  Returns:
    A Pandas Series containing the frequency of each word token.
  """

  words = []
  for i in range(len(df)):
    words.extend(df.loc[i, column_name].split())

  word_counts = pd.Series(words).value_counts()
  return word_counts
```

### Example of word_token counter function

```python
# Example usage:

df = pd.DataFrame({'text': ['This is a sample text', 'This is another sample text']})

# Tokenize and count the frequency of words in the 'text' column.
word_counts = tokenize_and_count_words(df, 'text')

# Print the word counts.
print(word_counts)
```

```python
import matplotlib.pyplot as plt
import pandas as pd

def generate_bar_chart(df, x_axis_column, y_axis_column):
  """Generates a bar chart with the given data frame.

  Args:
    df: A Pandas DataFrame.
    x_axis_column: The name of the column containing the X-axis data.
    y_axis_column: The name of the column containing the Y-axis data.
  """

  plt.figure(figsize=(20, 10))
  plt.bar(df[x_axis_column], df[y_axis_column], color='blue')
  plt.xlabel(x_axis_column)
  plt.ylabel(y_axis_column)
  plt.title('categories: word_token counts for all brands')
  plt.show()
```
